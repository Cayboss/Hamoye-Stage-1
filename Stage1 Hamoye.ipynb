```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading dataset
data_url = "https://github.com/HamoyeHQ/HDSC-Introduction-to-Python-for-machine-learning/files/7768140/FoodBalanceSheets_E_Africa_NOFLAG.csv"
data = pd.read_csv(data_url, encoding = "latin")
data.head()
```

```python
#Total sum of animal fat produced in 2014 and 2017
q11 = data.groupby("Item")["Y2014", "Y2017"].sum()
q11
```

```python
#Mean and std.d in 2015
q12_mean = round(data.Y2015.mean(), 3)
q12_std = round(data.Y2015.std(), 3)

print(q12_mean, q12_std)
```

```python
#No of missing values and percentage in 2016
q13_total = len(data.Y2016)
q13_missing = data.Y2016.isna().sum()
q13_missingperc = round((q13_missing/q13_total)*100, 2)
print(q13_missing, q13_missingperc)
```

```python
#Highest crrelation with element code
q14 = data.corr()
q14
```

```python
#Year with the highest sum of import quantity

#Check columns
columns = list(data.columns)

#Create a list to hold the years columns
years = []

#Create a loop to select the columns
for col in columns:
    if col.startswith("Y"):
        years.append(col)
    else:
        continue

#Using groupby to get the selected columns
q15 = data.groupby(by = [data["Element"] == "Import Quantity"])[years].sum()

print(q15)
print()
locs = q15.values[1].max()

print(locs)
```

```python
#Sum of production in 2014
q16 = data.groupby("Element")["Y2014"].sum()
q16
```

```python
#Highest sum by element in 2018 
q17 = data.groupby("Element")["Element", "Y2018"].sum().sort_values(by="Y2018", ascending=False)
q17[:5]
```

```python
#Third lowest element in 2018
q18 = q17.sort_values(by="Y2018", ascending=True)
q18[:5]
```

```python
#Total import quantity (Algeria) for 2018
q19 = data.groupby(by=["Area", "Element"])["Y2018"].sum()
q19[:10]
```
